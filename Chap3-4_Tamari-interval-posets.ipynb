## Contacts, descents and complement involution

```python
tip = TamariIntervalPoset(8, [(1, 6), (2, 6), (3, 5), (4, 5), (5, 6), (6, 8), (7, 8), (7, 6), (4, 3), (3, 1), (2, 1)])
tip
```

```python
viewLatex(tip)
```

```python
viewLatex(tip.lower_dyck_word())
```

```python
viewLatex(tip.upper_dyck_word())
```

```python
contact_vector = [len(tip.decreasing_roots())] + [len(tip.decreasing_children(i)) for i in range(1,8)]
contact_vector
```

```python
descent_vector = [len(tip.increasing_roots())] + [len(tip.increasing_children(i)) for i in range(8,1,-1)]
descent_vector
```

```python
tip_comp = tip.complement()
viewLatex(tip_comp)
```

## Rises and left-branch involution

```python
tip = TamariIntervalPoset(8, [(1,2), (2,4), (3,4), (6,7), (3,2), (5,4), (6,4), (8,7)])
viewLatex(tip)
```

```python
graftree = tip.grafting_tree()
viewLatex(graftree)
```

```python
leftbranch_gt = graftree.left_border_symmetry()
viewLatex(leftbranch_gt)
```

```python
leftbranch_tip = TamariIntervalPosets.from_grafting_tree(leftbranch_gt)
viewLatex(leftbranch_tip)
```
