## The weak order

```python
P3 = Permutations(3)
W3_left = Poset((P3, lambda x,y: x.permutohedron_lequal(y, side="left")))
W3_left
```

```python
W3_right = Poset((P3, lambda x,y: x.permutohedron_lequal(y, side="right")))
W3_right
```

```python
P4 = Permutations(4)
W4_right = Poset((P4, lambda x,y: x.permutohedron_lequal(y, side="right")))
W4_right.plot()
```

```python
W4_right.is_lattice()
```

```python
p1 = Permutation([3,1,4,2])
```

```python
# permutations smaller than p1 in the right weak order
p1.permutohedron_smaller()
```

```python
# permutations smaller than p1 in the left weak order
p1.permutohedron_smaller(side = "left")
```

```python
p2 = Permutation([3,2,1,4])
```

```python
# the meet and join of p1 and p2 in right weak order
p1.permutohedron_meet(p2)
```

```python
p1.permutohedron_join(p2)
```

```python
# the meet and join of p1 and p2 in left weak order
p1.permutohedron_meet(p2, side = "left")
```

```python
p1.permutohedron_join(p2, side = "left")
```
